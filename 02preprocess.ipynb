### GBIF

```python
local_file = "gbif"
!ls gbif/
```

```python
files = [p for p in pathlib.Path(local_file).rglob("*.json")]
results = []
for f in files:
    print(f)
    with open(f, "r") as arquivo:
        data = json.load(arquivo)
        if data and "results" in data and len(data["results"]) > 0:
            results.append(data["results"])
print("qtd em results: %d" % len(results))
```

```python
data = list(itertools.chain(*(r for r in results)))
d2= dict(itertools.chain.from_iterable(d.items() for d in data))
df = pd.json_normalize(data)
df.head()
```

```python
filename = os.path.join(local_file, "Piper.parquet")
df.to_parquet(filename)
!ls gbif/
```
